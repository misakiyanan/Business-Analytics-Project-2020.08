```python
"""
set reordered column in order_product table
"""
```

```python
import pandas as pd
import numpy as np
```

```python
order_products = pd.read_csv("data/order_products.csv")
orders = pd.read_csv("data/orders.csv")
joint = order_products.merge(orders, how="left", on="order_id")
# joint = joint.iloc[:,0:5]
joint
```

```python
grouped = joint.groupby(by=['user_id', 'product_id'])
joint['rank'] = grouped['order_number'].rank(ascending=True)
joint

# ref. https://stackoverflow.com/questions/26720916/faster-way-to-rank-rows-in-subgroups-in-pandas-dataframe
```

```python
joint['reordered'] = 0
joint
```

```python
joint.loc[joint['rank'] > 1, 'reordered'] = 1   
# ref.https://stackoverflow.com/questions/49161120/pandas-python-set-value-of-one-column-based-on-value-in-another-column
joint
```

```python
zero = joint[joint["reordered"] == 0]
one = joint[joint["reordered"] == 1]
print("number of reordered==0: ", zero.shape[0])
print("number of reordered==1: ", one.shape[0])
```

```python
df = pd.read_csv("data/order_products_linkedresult/order_products.csv")
zeros = df[df["reordered"] == 0]
ones = df[df["reordered"] == 1]
print("number of reordered==0: ", zeros.shape[0])
print("number of reordered==1: ", ones.shape[0])
```
